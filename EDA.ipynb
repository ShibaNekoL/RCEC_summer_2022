Groupby date (`yyyymmdd`) to visualize

```python
df_mean = df_cleaned.groupby(['yyyymmdd'])['TX01', 'PP01', 'WD01', 'WD02', 'PS01'].mean()
```

```python
def plot_mean(y):
    # Plot
    fig, ax = plt.subplots(1, 1, figsize = (100,3))

    plt.plot(df_mean[y])

    # Labelling 
    step = 1
    ax.xaxis.set_major_locator(md.YearLocator(step, month = 1, day = 1))
    ax.xaxis.set_major_formatter(md.DateFormatter('%Y'))

    plt.xticks(rotation=45)
    plt.xlabel("Date")
    plt.ylabel(y)
    plt.title(y + "Time Series Plot")

    # Display
    plt.show()
```

```python
for y in ['TX01', 'PP01', 'WD01', 'WD02', 'PS01']:
    plot_mean(y)
```
