## Station position data

```python
st_df = pd.read_csv('stations.csv')
st_df
```

```python
st_df_for_merge = st_df[['站號','站名','海拔高度(m)','經度','緯度']]
st_df_for_merge
```

## Merge

```python
def taiwan_stations_onevar(df, var):
    df_right = df.loc[:, df.columns.str.startswith(var)]
    df_right = df_right.transpose()
    df_right.index = df_right.index.str.lstrip(var + '_')
    df_right = pd.DataFrame(df_right)
    return st_df_for_merge.merge(df_right, left_on='站號', right_index=True, how='right')
```

```python
taiwan_stations_onevar(cluster_centers_projected_back, 'PP01')
```
